response.status_code :

200: OK - Yêu cầu đã thành công. Mã trạng thái này thường xuất hiện khi bạn nhận được nội dung thành công từ máy chủ.

201: Created - Yêu cầu đã tạo thành công một tài nguyên mới trên máy chủ.

204: No Content - Yêu cầu đã thành công, nhưng không có nội dung được trả về.

400: Bad Request - Yêu cầu không hợp lệ hoặc có lỗi cú pháp.

401: Unauthorized - Yêu cầu yêu cầu xác thực. Người dùng chưa được ủy quyền truy cập.

403: Forbidden - Yêu cầu bị từ chối vì người dùng không có quyền truy cập tài nguyên.

404: Not Found - Tài nguyên hoặc trang web không được tìm thấy.

500: Internal Server Error - Lỗi nội bộ trên máy chủ, thường là do lỗi phía máy chủ.

```python
import requests
#gửi yêu cầu GET đến URL
response = requests.get("https://api.tiki.vn/aerith/v1/p")
#thông báo trang thái khi gửi yêu cầu đến URL
response.status_code
#lưu nội dung với dạng bytes
response.content
#lưu nội dung văn bản
response.text
response.headers
#Thể hiện giá trị của nó
response.headers['Content-Type']
```
